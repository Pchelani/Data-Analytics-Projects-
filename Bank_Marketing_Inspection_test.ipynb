# Bank Marketing Data Analysis
This notebook presents an analysis of the `bankmarketing.csv` dataset, which includes data related to a bank's marketing campaigns. The main goal is to understand customer behavior and predict whether a client will subscribe to a term deposit.

```python

import pandas as pd

# Load the dataset
df = pd.read_csv('bankmarketing.csv')

# Display the first few rows
df.head()
df = pd.read_csv("bankmarketing.csv")

# 1. First 5 rows
print("\n----- First 5 Rows -----")
display(df.head())

# 2. Shape of dataset
print("\n----- Dataset Shape -----")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")

# 3. Column info (data types + non-null counts)
print("\n----- Dataset Info -----")
print(df.info())
```

```python

# Check for missing values and data types
print("\n----- Missing Values -----")
print(df.isnull().sum())

```

```python

# Summary statistics
print("\n----- Summary Statistics -----")
display(df.describe(include="all"))
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Plot the distribution of the target variable
# Target column
target_col = 'y'

# Count values
counts = df[target_col].value_counts()

# Plot with updated size (8,6)
plt.figure(figsize=(8,6))
plt.bar(counts.index, counts.values)
plt.xlabel('Target Variable (y)')
plt.ylabel('Count')
plt.title('Distribution of Target Variable')
plt.tight_layout()
plt.show()
```

```python

# Correlation matrix for numerical features
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

print("Numerical Columns:")
print(list(num_cols))

# Plot histograms for all numerical columns
df[num_cols].hist(figsize=(15, 12), bins=30)
plt.tight_layout()
plt.show()
```
